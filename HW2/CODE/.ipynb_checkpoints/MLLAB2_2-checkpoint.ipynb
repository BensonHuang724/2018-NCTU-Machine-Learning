```python
predict=np.zeros((180,3))
target=np.zeros((180,3))
for i in range (0,180):
    for j in range (0,3):
        target[i][j]=train.iloc[i][j]
```

# function

```python
####################################################################################################
def update_predict(predict,w,train):
    for i in range (0,180):
        numerator=np.zeros((1,3))
        for j in range (0,3):
            for k in range (0,7):
                numerator[0][j]=numerator[0][j]+w[j][k]*train.iloc[i][k+3]
        ##print("debug:",numerator)        
        denominator=0
        print()
        for j in range (0,3):
            denominator=denominator+np.exp(numerator[0][j])
        
        for j in range (0,3):
            predict[i][j]=np.exp(numerator[0][j])/denominator
        
    return predict
####################################################################################################
def Error_f(predict,target):
    error=0
    for i in range (0,180):
        for j in range (0,3):
            error=error+target[i][j]*np.log(predict[i][j])
            
    return -1*error
####################################################################################################
def d_Error_f(j,predict,target,train_data):
    d_error=np.zeros((1,7))
    basis=np.zeros((1,7))
    for i in range (0,180):
##################################################construct basis
        for k in range (0,7):
            basis[0][k]=train_data.iloc[i][k+3]
##################################################construct basis
            d_error=d_error+ (predict[i][j]-target[i][j] )*basis
    return d_error
####################################################################################################
def hessian(j,predict,train_data):
    H=np.zeros((7,7))
    basis=np.zeros((1,7))
    for i in range (0,180):
##################################################construct basis
        for k in range (0,7):
            basis[0][k]=train_data.iloc[i][k+3]
##################################################construct basis
        ##print(basis)
        H=H+(predict[i][j])*(1-predict[i][j])*basis.T.dot(basis)
        ##print("H:",H,"")
    return H
####################################################################################################
def cal_train_acc(train,predict):
    predict_class=np.zeros((180,1))
    for i in range(0,180):
        if (predict[i][0]>=predict[i][1]) and ((predict[i][0]>=predict[i][2])):
            predict_class[i][0]=1
        elif (predict[i][1]>=predict[i][0]) and ((predict[i][1]>=predict[i][2])):
            predict_class[i][0]=2
        elif (predict[i][2]>=predict[i][0]) and ((predict[i][2]>=predict[i][1])):
            predict_class[i][0]=3
    correct=0
    ##print (predict_class)
    for i in range(0,180):
        if (train.loc[i][10]==predict_class[i][0]):
            correct=correct+1
    return correct/180*100
            
    
```

# 2_1 Newton's method (1 iteration)

```python
update_predict(predict,w,train)
```

```python
error=10000000000000
iteration=1
while error>100:
    ##w_new0=np.zeros((1,7))
    ##w_new1=np.zeros((1,7))
    ##w_new2=np.zeros((1,7))
    w_old0=w[0][:].copy()
    w_old1=w[1][:].copy()
    w_old2=w[2][:].copy()
    
    w[0][:] = w_old0-d_Error_f(0,predict,target,train).dot( np.linalg.pinv(hessian(0,predict,train).T) )
    w[1][:] = w_old1-d_Error_f(1,predict,target,train).dot( np.linalg.pinv(hessian(1,predict,train).T) )
    w[2][:] = w_old2-d_Error_f(2,predict,target,train).dot( np.linalg.pinv(hessian(2,predict,train).T) )
    
    predict=update_predict(predict,w,train)
    error=Error_f(predict,target)
    print("iteration:",iteration,"  error:",error)
    iteration=iteration+1
    print(w)
print(error)   
    
```

```python
input_dimension=7
num_class3=60
num_class2=60
num_class1=60
u=np.zeros((1,input_dimension))
u1=np.zeros((1,input_dimension))
u2=np.zeros((1,input_dimension))
u3=np.zeros((1,input_dimension))
temp=np.zeros((1,input_dimension))
###########################################
for j in range(0,num_class1):
    for k in range(0,input_dimension):
        temp[0][k]=train.iloc[j][k+3]
    u1=u1+temp;
u1=u1/num_class1
###########################################
for j in range(num_class1,num_class1+num_class2):
    for k in range(0,input_dimension):
        temp[0][k]=train.iloc[j][k+3]
    u2=u2+temp;
u2=u2/num_class2
###########################################
for j in range(num_class1+num_class2,num_class1+num_class2+num_class3):
    for k in range(0,input_dimension):
        temp[0][k]=train.iloc[j][k+3]
    u3=u3+temp;
u3=u3/num_class3
###########################################
u=(num_class1*u1+num_class2*u2+num_class3*u3)/(num_class1+num_class2+num_class3)    
###########################################
u
```

```python
Sw=np.zeros((input_dimension,input_dimension))
for j in range(0,num_class1):
    for k in range(0,input_dimension):
        temp[0][k]=train.iloc[j][k+3]
    Sw=Sw+(temp.T).dot(temp)
    
for j in range(num_class1,num_class1+num_class2):
    for k in range(0,input_dimension):
        temp[0][k]=train.iloc[j][k+3] 
    Sw=Sw+(temp.T).dot(temp)
for j in range(num_class1+num_class2,num_class1+num_class2+num_class3):
    for k in range(0,input_dimension):
        temp[0][k]=train.iloc[j][k+3]
    Sw=Sw+(temp.T).dot(temp)
Sw
```

```python
SB=np.zeros((input_dimension,input_dimension))
SB=num_class1* ( (u1-u).T).dot((u1-u)) +num_class2*((u2-u).T).dot((u2-u))+num_class3*((u3-u).T).dot((u3-u))
SB
```
