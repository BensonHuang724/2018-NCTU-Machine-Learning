```python
stopping_criterion=100
w=np.zeros((3,2))
num_train=train.size/10
num_test = test.size/7
w
```

```python
predict=np.zeros((180,3))
target=np.zeros((180,3))
for i in range (0,180):
    for j in range (0,3):
        target[i][j]=train.iloc[i][j]
```

```python
def update_predict(predict,w,train):
    for i in range (0,180):
        numerator=np.zeros((1,3))
        for j in range (0,3):
            for k in range (0,2):
                numerator[0][j]=numerator[0][j]+w[j][k]*train.iloc[i][k+3]     
        denominator=0
        for j in range (0,3):
            denominator=denominator+np.exp(numerator[0][j])
        
        for j in range (0,3):
            predict[i][j]=np.exp(numerator[0][j])/denominator
        
    return predict

def Error_f(predict,target):
    error=0
    for i in range (0,180):
        for j in range (0,3):
            error=error+target[i][j]*np.log(predict[i][j])
            
    return -1*error

def d_Error_f(j,predict,target,train_data):
    d_error=np.zeros((1,2))
    basis=np.zeros((1,2))
    for i in range (0,180):
        for k in range (0,2):
            basis[0][k]=train_data.iloc[i][k+3]
            d_error=d_error+ (predict[i][j]-target[i][j] )*basis
    return d_error

def hessian(j,predict,train_data):
    H=np.zeros((2,2))
    basis=np.zeros((1,2))
    for i in range (0,180):
        for k in range (0,2):
            basis[0][k]=train_data.iloc[i][k+3]
        H=H+(predict[i][j])*(1-predict[i][j])*basis.T.dot(basis)
    return H

def cal_train_acc(train,predict):
    predict_class=np.zeros((180,1))
    for i in range(0,180):
        if (predict[i][0]>=predict[i][1]) and ((predict[i][0]>=predict[i][2])):
            predict_class[i][0]=1
        elif (predict[i][1]>=predict[i][0]) and ((predict[i][1]>=predict[i][2])):
            predict_class[i][0]=2
        elif (predict[i][2]>=predict[i][0]) and ((predict[i][2]>=predict[i][1])):
            predict_class[i][0]=3
    correct=0
    for i in range(0,180):
        if (train.loc[i][10]==predict_class[i][0]):
            correct=correct+1
    return correct/180*100
```

```python
error=10000000000000
iteration=1
while error>100:
    ##w_new0=np.zeros((1,7))
    ##w_new1=np.zeros((1,7))
    ##w_new2=np.zeros((1,7))
    w_old0=w[0][:].copy()
    w_old1=w[1][:].copy()
    w_old2=w[2][:].copy()
    
    w[0][:] = w_old0-d_Error_f(0,predict,target,train).dot( np.linalg.pinv(hessian(0,predict,train).T) )
    w[1][:] = w_old1-d_Error_f(1,predict,target,train).dot( np.linalg.pinv(hessian(1,predict,train).T) )
    w[2][:] = w_old2-d_Error_f(2,predict,target,train).dot( np.linalg.pinv(hessian(2,predict,train).T) )
    
    predict=update_predict(predict,w,train)
    error=Error_f(predict,target)
    print("iteration:",iteration,"  error:",error)
    iteration=iteration+1
    print(w)
print(error)   
```
